```python
#Import Libraries
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

#Load Images
image=cv2.imread("Downloads\\image.jpg")
plt.imshow(image)

#Display in openCV
#cv2.imshow("image1",image)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
#size of the image

image.shape[:3]
```

```python
#convert to gray scale
gray_img= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

```python
#convert to HSV scale
#Hue-(0,179), Saturation-(0-255), Value-(0-255)
hsv_img= cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_img)
```

```python
#split the image into each color
B,G,R=cv2.split(image)
#Blue scale
plt.imshow(B)
```

```python
#Green scale
plt.imshow(G)
```

```python
#Red scale
plt.imshow(R)
```

```python
#Merge the image
Merged=cv2.merge([B,G,R])
plt.imshow(Merged)
```

```python
#Amplify color
merged=cv2.merge([B+100,G,R])
#merged=cv2.merge([B,G+100,R]) Green amplifier
#merged=cv2.merge([B,G,R+100])  Red amplifier
plt.imshow(merged)
w=10
h=
```

```python
#Histogram
histogram=cv2.calcHist([image],[0],None,[256],[0,256])
plt.hist(image.ravel(),256,[0,256])
plt.show()
```

```python
# seperate color ranges
color=('B','G','R')
for i,col in enumerate(color):
    hist= cv2.calcHist([image],[1],None,[256],[0,256])
    plt.plot(hist, color="skyblue")
    plt.xlim([0,256])
plt.show()
```
